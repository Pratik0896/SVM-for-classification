Training a SVM model to classify human cells in benign or malignant.
dataset has class attribute with value=2 (benign) and value=4 (malignant) 

```python
#import libraries.
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
#import dataset.
data = pd.read_csv('cell_samples.csv')
data.head()
```

```python
#check data in target set (class attribute of dataset).
data[['Class']]
```

```python
list(data['Class'].unique())
```

```python
#check for basic statistics.
data.describe()
```

```python
#verifying datatypes
data.dtypes
```

```python
#Machine learning model cannot process object type data. therefore converting object type data into Numeric format.
data = data[pd.to_numeric(data['BareNuc'],errors='coerce').notnull()]
data['BareNuc'] = data['BareNuc'].astype('int64')
```

```python
data.dtypes
```

```python
#seleccting feature set and target set from the pre-processed dataset.
feature_set = data.iloc[:,1:10]
feature_set
```

```python
#selecting target set.
target_set = data.iloc[:,-1]
target_set
```

```python
#splitting into training and testing sets.
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(feature_set, target_set, test_size=0.2, random_state = 0) 
```

```python
#creating SVM model
from sklearn import svm
model = svm.SVC(kernel='rbf')
model.fit(train_x, train_y)
```

```python
#predicting the class variable(target_set).
y_pred = model.predict(test_x)     
y_pred
```

```python
#find accuracy using f1 score.
from sklearn.metrics import f1_score
accuracy = f1_score(test_y, y_pred, average='weighted')
```
